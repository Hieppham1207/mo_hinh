# MÔ TẢ TẬP DỮ LIỆU
Tập dữ liệu "Data_hospital.csv" chứa thông tin về 106,987 cuộc hẹn tại một bệnh viện, với mục tiêu phân tích và dự đoán khả năng bệnh nhân không đến cuộc hẹn. Dữ liệu bao gồm 15 cột, mỗi cột mô tả các thuộc tính khác nhau liên quan đến bệnh nhân và cuộc hẹn. Cụ thể:
---
* **Phần I. Thông tin về bệnh nhân:**
1. Mã bệnh nhân (PatientId): Mã định danh duy nhất cho từng bệnh nhân.
2. Giới tính (Gender): Giới tính của bệnh nhân (M:Nam/F:Nữ).
3. Tuổi (Age): Độ tuổi của bệnh nhân.

* **Phần II. Thông tin về sức khỏe:**
4. Tăng huyết áp (Hipertension): Tình trạng tăng huyết áp của bệnh nhân (1 là có, 0 là không).
5. Tiểu đường (Diabetes): Tình trạng bệnh tiểu đường (1 là có, 0 là không).
6. Nghiện rượu (Alcoholism): Tiền sử nghiện rượu (1 là có, 0 là không).
7. Khuyết tật (Handcap): Trạng thái khuyết tật của bệnh nhân (1 là có, 0 là không).


* **Phần III. Thông tin về khu vực:**
8. Khu vực dân cư (Neighbourhood): Nơi cư trú của bệnh nhân.

* **Phần IV. Thông tin về cuộc hẹn:**
9. Mã cuộc hẹn (AppointmentID): Mã định danh cho cuộc hẹn.
10. Ngày đặt lịch (ScheduledDay): Ngày mà bệnh nhân đã đặt lịch hẹn.
11. Ngày hẹn (AppointmentDay): Ngày thực tế cuộc hẹn diễn ra.
12. Chênh lệch ngày (Date.diff): Số ngày giữa ngày đặt lịch và ngày hẹn.


* **Phần V. Thông tin về hỗ trợ và nhắc nhở:**
13. Hỗ trợ học bổng (Scholarship): Bệnh nhân có nhận được trợ cấp y tế không (1 là có, 0 là không).
14. Đã nhận tin nhắn (SMS_received): Bệnh nhân có nhận được tin nhắn nhắc lịch hẹn không (1 là có, 0 là không).


* **Phần VI. Kết quả cuộc hẹn:**
15. Đã đến cuộc hẹn (Showed_up): Kết quả của cuộc hẹn (1 nếu bệnh nhân đến, 0 nếu không).

```python
group_huyet_ap = data.groupby(['Showed_up', 'Hipertension']).size().unstack(fill_value=0)
print(group_huyet_ap)

# Tính tỷ lệ phần trăm của người mắc bệnh huyết áp trong từng nhóm Showed_up
group_tl_huyet_ap = group_huyet_ap.div(group_huyet_ap.sum(axis=1), axis=0) * 100

print(group_tl_huyet_ap)
group_tl_huyet_ap = group_tl_huyet_ap.reset_index()
group_tl_huyet_ap = group_tl_huyet_ap.melt(id_vars=['Showed_up'], value_vars=[0, 1], var_name='Hipertension', value_name='Percentage')
group_tl_huyet_ap['Hipertension'] = group_tl_huyet_ap['Hipertension'].map({0: 'Không mắc bệnh', 1: 'Mắc bệnh'})

# Tạo biểu đồ cột tỷ lệ phần trăm người mắc bệnh huyết áp
plt.figure(figsize=(10, 6))
sns.barplot(data=group_tl_huyet_ap, x='Showed_up', y='Percentage', hue='Hipertension', palette='coolwarm')
plt.title('Tỷ lệ phần trăm người mắc bệnh huyết áp theo việc có mặt')
plt.xlabel('Có mặt')
plt.ylabel('Tỷ lệ phần trăm (%)')
plt.xticks(ticks=[0, 1], labels=['Chưa đến', 'Đã đến'])
plt.legend(title='Bệnh huyết áp')
plt.show()
```

# Nhận xét:
* 80% trong số những người không đến cuộc hẹn là không mắc bệnh huyết áp (màu xanh).
* Chỉ có khoảng 20% là người mắc bệnh huyết áp (màu cam)
* Trong số những người đã đến cuộc hẹn, khoảng 25% là người mắc bệnh huyết áp (màu cam).
* Khoảng 75% là người không mắc bệnh huyết áp (màu xanh)

*Những người mắc bệnh huyết áp có xu hướng tham gia các cuộc hẹn y tế hơn, có thể do tình trạng sức khỏe của họ đòi hỏi sự theo dõi cẩn thận hơn.
*Ngược lại, những người không mắc bệnh huyết áp có xu hướng bỏ lỡ các cuộc hẹn nhiều hơn, điều này có thể là do họ cảm thấy ít cần thiết hơn trong việc phải tuân thủ các cuộc hẹn y tế.

```python
# Import thư viện cần thiết


from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report,confusion_matrix

# Chia dữ liệu thành tập huấn luyện và kiểm tra
X = data[['Hipertension', 'Diabetes', 'Alcoholism', 'Handcap']]
y = data['Showed_up']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Tăng cường lớp thiểu số bằng SMOTE
smote = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)

# Huấn luyện mô hình với dữ liệu đã resample
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train_resampled, y_train_resampled)

# Dự đoán trên tập kiểm tra
y_pred = model.predict(X_test)

# Đánh giá mô hình
print(classification_report(y_test, y_pred))


# Vẽ confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```
